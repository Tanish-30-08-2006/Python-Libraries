```python
import numpy as np
```

```python
#-----------------------2D ARRAY INDEXING--------------#

arr = np.arange(1,31)
arr = arr.reshape(6,5)
print(arr)
```

```python
# -----ROWS-------#
print(arr[0])
print(arr[1])
print(arr[2])
print(arr[3])
print(arr[4])
print(arr[5])
```

```python
#--------ELEMENTS------#
print(arr[5][4])
```

```python
# ---------------------2D ARRAY SLICING---------------------#


print(arr)

# we want 2,3,7,8 square to be sliced out

print(arr[0:2,1:3]) # 0,1 two rows to slice so 0 to 2 slicing for rows and 1,2 two columns to slice so 1 to 3 column slicing

# we want 19-30 matrix sliced out (3*2)
print("\n")
print(arr[3:6,3:5])

print("\n")
# to go to end by default leave end index empty 
print(arr[3:,3:])
```

```python
# ------------TO SELECT COLUMNS AND ROWS------------#

print(arr)

print("\n")
print(arr[0,:])

print("\n")
print(arr[:,2:4])

print("\n")
print(arr[:,4])

```
