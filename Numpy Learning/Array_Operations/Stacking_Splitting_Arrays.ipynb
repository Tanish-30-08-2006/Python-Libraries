```python
import numpy as np
```

```python

arr1 = np.arange(1,5)
arr2 = np.arange(5,9)

print(arr1)
print(arr2)
```

```python
#------------STACKING ARRAY---------------#

# Note we CAN NOT stack 2D ARRAYS we need 1D ARRAYS (VECTORS) to stack

print(np.vstack((arr1,arr2))) # VERTICAL STACK 
print("\n")

print(np.hstack((arr1,arr2))) # HORIZONTAL STACK 
print("\n")

print(np.column_stack((arr1,arr2))) # COLUMN STACK
```

```python
#_--------------SPLITTING ARRAY-----------------#

arr = np.arange(16).reshape(4,4)
print(arr)
print("\n")

print(np.hsplit(arr,4))

print("\n")
print(np.vsplit(arr,4))

print("\n")
print(np.vsplit(arr,2))

# NOTE: SPLITTED ARRAYS ARE ALSO 2D ARRAYS EVEN WHEN WE BREAK IT IN 4 PARTS 
```
