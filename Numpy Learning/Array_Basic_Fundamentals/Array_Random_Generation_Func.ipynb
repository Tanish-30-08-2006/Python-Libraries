```python
import numpy as np
import random #inbuilt library of python
```

```python

#--------------GENERATING N RANDOM NUMBERS(DECIMALS) -------------#

#----------NORMALIZATION---------#
arr = np.random.rand(5)
print(arr)

arr = np.random.rand(2)
print(arr)

#----------STANDARDIZATION--------#
arr = np.random.randn(5)
print(arr)

arr = np.random.randn(2)
print(arr)

# point to note every random number shown will be between 0 and 1 in first part and -3 and 3 in second 
# this is due to cocnept called NORMALIZATION in statistics and when valuesare between 3 and -3 it is called STANDARDIZATION
```

```python
#----------------GENERATING RANDOM INTEGERS-----------#

# generates 1 random integer between 0 and 6
num = np.random.randint(6)
print(num)

# to create array of random numbers
arr = np.random.randint(1,10,5,int) # between 1 and 10 r random integers
print(arr)
```
